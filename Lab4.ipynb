```python
# Task 1 Identify data quality issues in the dataset

# The dataset exhibits several common data quality challenges
# that need addressing before further analysis.

# There are missing values in key numerical columns such as LotFrontage, GarageYrBlt, and MasVnrArea,
# which require appropriate imputation or handling.

# Outliers are present, notably in variables like LotFrontage and GarageYrBlt,
# which could skew results if left untreated.

# Some categorical variables suffer from inconsistency in labeling.

# Additionally, certain columns have inappropriate data types that must be corrected,
# such as treating numeric codes like MSSubClass as categorical variables.

# Lastly, there are occasional illogical or invalid values,
# such as construction years beyond the current year,
# which should be reviewed and cleaned.
```

```python
import pandas
import numpy
import matplotlib.pyplot
import seaborn
seaborn.set(style="whitegrid")
```

```python
pandas.set_option("display.max_columns", None)
pandas.set_option('display.max_rows', None)
tr = pandas.read_csv('C:\\Users\\Zahra\\Downloads\\Htrain.csv')
```

```python
tr.head()
```

```python
tr.dtypes
```

```python
# 1. MSSubClass
tr['MSSubClass'] = tr['MSSubClass'].astype('category')

# 2. Yes/No
tr['CentralAir'] = tr['CentralAir'].map({'Y':1,'N':0})

# 3. Quality
qual_map = {"Ex":5,"Gd":4,"TA":3,"Fa":2,"Po":1}

cols = ['ExterQual','ExterCond','BsmtQual','BsmtCond',
        'HeatingQC','KitchenQual','FireplaceQu',
        'GarageQual','GarageCond','PoolQC']

for col in cols:
    tr[col] = tr[col].map(qual_map)

# 4. Fill NA
cols_na = ['Alley','BsmtQual','BsmtCond','BsmtExposure',
           'BsmtFinType1','BsmtFinType2','FireplaceQu',
           'GarageType','GarageFinish','GarageQual',
           'GarageCond','PoolQC','Fence','MiscFeature']

for col in cols_na:
    tr[col] = tr[col].fillna('None')

# 5. Convert text → category
cat_cols = tr.select_dtypes(include=['object', 'string']).columns

for col in cat_cols:
    tr[col] = tr[col].astype('category')
```

```python
print(tr.isna().sum())
```

```python
# Task 2 Apply one missing value strategy and explain why.

# In my case, the best approach for handling missing values is to
# use median imputation for numerical columns and "None" for categorical.
# This is because some numerical columns like LotFrontage, MasVnrArea, and GarageYrBlt
# have missing values, and using the median is robust against outliers
# and preserves the overall distribution of the data.
# 
# For categorical columns such as MasVnrType and Electrical,
# filling missing values with "None" provides a valid placeholder
# without losing data, making it simpler and safer
# than removing records or using the mean.
    
copy = tr.copy()

# Numeric columns – fill missing with median
copy['LotFrontage'] = copy['LotFrontage'].fillna(copy['LotFrontage'].median())
copy['MasVnrArea'] = copy['MasVnrArea'].fillna(copy['MasVnrArea'].median())
copy['GarageYrBlt'] = copy['GarageYrBlt'].fillna(copy['GarageYrBlt'].median())

# Categorical columns – add 'None' to categories first if necessary, then fillna
copy['MasVnrType'] = copy['MasVnrType'].cat.add_categories('None').fillna('None')
copy['Electrical'] = copy['Electrical'].cat.add_categories('None').fillna('None')

copy.head(10)
```

```python
# Task 3 Detect and handle outlers using IQR
matplotlib.pyplot.figure(figsize=(6,4))
seaborn.boxenplot(x=tr['LotFrontage'])
matplotlib.pyplot.title("Boxplot")
matplotlib.pyplot.show()
```

```python
Q1 = tr['LotFrontage'].quantile(0.25)
Q3 = tr['LotFrontage'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers_lotfrontage = tr[(tr['LotFrontage'] < lower) | (tr['LotFrontage'] > upper)]
print("Outliers in LotFrontage:")
outliers_lotfrontage.head()
```

```python
Q1 = tr['MasVnrArea'].quantile(0.25)
Q3 = tr['MasVnrArea'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers_masvnrarea = tr[(tr['MasVnrArea'] < lower) | (tr['MasVnrArea'] > upper)]
print("Outliers in MasVnrArea:")
outliers_masvnrarea.head()
```

```python
Q1 = tr['GarageYrBlt'].quantile(0.25)
Q3 = tr['GarageYrBlt'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers_garageYrBlt = tr[(tr['GarageYrBlt'] < lower) | (tr['GarageYrBlt'] > upper)]
print("Outliers in GarageYrBlt:")
outliers_garageYrBlt.head()
```

```python
print(tr['GarageYrBlt'].min(), tr['GarageYrBlt'].max())
print(lower, upper)

# The min and max values of 'GarageYrBlt' are 1900.0 and 2010.0
# The IQR-based lower and upper bounds are 1899.5 and 2063.5
# Since all values are within these bounds, there are no outliers in this column
```
